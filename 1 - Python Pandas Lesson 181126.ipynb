### Basic Data Cleanup
For now lets narrow down this dataset to some simple components:
- REGION
- YEAR
- WEEK
- ILITOTAL

````python
# we'll select just those columns and create new dataframe
df1 = df[['YEAR', 'WEEK','REGION','ILITOTAL']]
````

````python
# let's check the summary of our new dataframe using the "info()" function
df1.info()
````

Note that our "ILITOTAL" variable is an 'object' type, not an integer ('int64').  This means that there are some string values in this column that we should clean up.

````python
# let's take a closer look at our new dataframe
df1.head(10)
````

````python
# note that in row 9 we have a value of "X" for 'ILITOTAL'
# lets replace all the "X" values in 'ILITOTAL' with a null placeholder
# to do this we're going to import another pachage called NumPy
import numpy as np
df1['ILITOTAL'].replace('X', np.nan, inplace=True)
````

````python
df1.head(10)
````

````python
# lets
df1[.info()]
````

````python
df1['ILITOTAL'] = df1['ILITOTAL'].astype('float')
````

````python
# now lets recheck the info for our dataframe
# note: 'ILITOTAL' is now a float object ('float64')
df1.info()
````

````python
# lets try the "describe()" function again
# note: we now have summary states for 'ILITOTAL'
df1.describe()
````

````python
# one more safetey check, is this just for the 50 US states?
df1['REGION'].nunique()
````

````python
df1['REGION'].unique().tolist()
````

````python
# we'll create of list of the regions we'd like to remove
remove_regions = ['Puerto Rico', 'Virgin Islands', 'New York City']
````

````python
# now we'll select all the rows that do not include the regions in 'remove_regions'
# note: what happens if you remove the "~"?
df2 = df1[~df1['REGION'].isin(remove_regions)]
````

````python
df2.head()
````

### Working with time series
Pandas can be very helpful when trying to organize time-series data.  This is going to get a bit hairy, but we're going to parse down our data even further by lumping our "YEAR" and "WEEK" data into a single variable.

````python
# 
df2['date'] = pd.to_datetime(df2.YEAR.astype(str), format='%Y') + \
             pd.to_timedelta(df2.WEEK.mul(7).astype(str) + ' days')
````

````python
# we can check for our new column
df2.head()
````

````python
df2.info()
````

````python
df2.set_index('date',drop=True, append=False, inplace=True)
````
